## Importing The Seaborn Library To Help Create Interactive Plots to Help Draw Meaningful Infromation From The Dataset

```python
import seaborn as sns
```

# Quality Distribution

```python
df = pd.read_csv('red_wine.csv')  
```

```python
plt.figure(figsize=(5,5)) 
sns.displot(df['quality'], kde=False, bins=6)
plt.title('Distribution of Wine Quality Scores') 
plt.xlabel('Quality Rating')
plt.ylabel('Count')
plt.show()
```

# Quality VS Alcohol Boxplot

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='quality',y='alcohol',data = df, palette='dark:blue',hue = 'quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol (%)')
plt.title('Wine Quality vs Alcohol Content')
plt.show()
```

# Pairplot

```python
sns.pairplot(df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'quality',]])
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

# Correlation HeatMap

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

## Correlation Heatmap
### Overall Correlation:
- The heat map indicates that there is a significant correlation between all variables involved, the darker the color, the higher the correlation.
- White to blue represents negative correlation. As the color gets lighter, the absolute value of the correlation gets smaller.
- White to orange represents correlation near zero. White means no correlation, while orange shows correlation so small that it is not practically significant.
- Orange to red represents positive correlation. As the color gets darker, the absolute value of the correlation gets larger.

## Pairplots Of Key Features
- The pairplot builds on two basic figures, the histogram and the scatter plot. The histogram on the diagonal allows us to see the distribution of a single variable while the scatter plots on the upper and lower triangles show the relationship (or lack thereof) between two variables.For example, the left-most plot in the second row shows the scatter plot of volatile acidity versus fixed acidity.
- The default pairplot by itself often gives us valuable insights. We see that fixed acidity and citric acid are positively correlated showing that citric acid leads to increase in fixed acidity (although this of course does not prove that one causes the other). It also appears that citric acid volatile acidity are negatively correlated showing that and increase in citric acid leads to a decrease in volatile acidity.
